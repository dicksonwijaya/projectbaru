```python

import requests
from bs4 import BeautifulSoup

```

```python
search_query = input("请输入要搜索的YouTuber姓名、频道或称呼：")
search_url = f"https://news.tvbs.com.tw/news/searchresult/{search_query}/news"
response = requests.get(search_url)
soup = BeautifulSoup(response.text, "html.parser")


main_tag = soup.find('main')
elements_to_exclude = main_tag.select('.keyword, .search, .page-div, .sidebar_div')


for element in elements_to_exclude:
    element.extract()


li_elements = main_tag.find_all('li')


for li in li_elements:

    text = li.get_text()
    a_tag = li.find('a')

    if a_tag:
        link = a_tag.get('href')  
    else:
        link = 'No link found' 

    # 提取图像链接
    img_tag = li.find('img')  
    if img_tag:
        img_src = img_tag.get('data-original')  
    else:
        img_src = 'No image found'  

    print("Text:", text)
    print("Link:", link)
    print("Image URL:", img_src)
    print("\n")
```

```python

start_page = 1
search_query = input("请输入要搜索的YouTuber姓名、频道或称呼：")
base_url = f"https://news.tvbs.com.tw/news/searchresult/{search_query}/news/"
while True:

    current_url = f"{base_url}{start_page}"

    response = requests.get(current_url)
    soup = BeautifulSoup(response.text, "html.parser")

    main_tag = soup.find('main')

    elements_to_exclude = main_tag.select('.keyword, .search, .page-div, .sidebar_div')
    for element in elements_to_exclude:
        element.extract()

    
    li_elements = main_tag.find_all('li')


    for li in li_elements:
       
        text = li.get_text()

       
        a_tag = li.find('a')
        

        if a_tag:
            link = a_tag.get('href')  
        else:
            link = 'No link found'  

        # 提取图像链接
        img_tag = li.find('img') 
        if img_tag:
            img_src = img_tag.get('data-original') 
        else:
            img_src = 'No image found'  


        print("Text:", text)
        print("Link:", link)
        print("Image URL:", img_src)
        print("\n")


    current_page_element = soup.find('a', class_='page-box active')
    if current_page_element:
        current_page = int(current_page_element.get_text())
    else:
        break  


    next_page_element = soup.find('a', class_='page-box').find_next_sibling('a', class_='page-box')
    
   
    if next_page_element:
        start_page += 1
    else:
        break  
```
